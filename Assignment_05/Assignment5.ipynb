```python
%matplotlib inline
from sklearn import svm
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

## Read the Dataset

```python
data = loadmat('ex6data1.mat')
X = data['X']
y = data['y']
```

# Split dataset for train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Plot the data

```python
from matplotlib.colors import ListedColormap
y_test_ = 'k'
def plotData():
    plt.figure(figsize=(10,7))
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=ListedColormap(['r', 'b']))
    plt.scatter(X_test[:, 0],  X_test[:, 1],  c=y_test_, cmap=ListedColormap(['r', 'b']))
    plt.show()
    
plotData()
```

#### Create and Visualize kNN

```python
def kNN(k=3): 
    kNN = []
    for x in X_test:
        euclidean_distance = np.sqrt(np.sum((X_train - x)**2, axis=1))
        __, y_train_sorted = zip(*sorted(zip(np.array([euclidean_distance]).T, y_train)))
        freq1=0; freq2=0
        for d in y_train_sorted[:k]: 
            if   d == 0: freq1 += 1
            elif d == 1: freq2 += 1
        kNN.append(0 if freq1 > freq2 else 1)
    return np.array([kNN]).T

y_test_ = kNN()
plotData()
```

## Compute accuracy

```python
print(f"kNN accuracy: {np.mean(y_test == y_test_)*100:.1f}%")
```

## Create LinearSVC and train your data

```python
clf = svm.LinearSVC(C=1e10, dual=False)
clf.fit(X_train, y_train.ravel())
pred = np.array([clf.predict(X_test)]).T
```

## Compute accuracy

```python
print(f"SVM accuracy: {clf.score(X_test, y_test)*100:.1f}%")
```

## Visualize the Decision Boundary

```python
def plotDecisionBoundary():
    xx, yy = np.meshgrid(
        np.linspace(X[:, 0].min()-0.2, X[:, 0].max()+0.2),
        np.linspace(X[:, 1].min()-0.2, X[:, 1].max()+0.2)
    )
    zz = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    plt.figure(figsize=(10,7))
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, s=40, cmap=ListedColormap(['r', 'b']))
    plt.scatter(X_test[:, 0],  X_test[:, 1],  c=y_test_, s=40, cmap=ListedColormap(['r', 'b']))
    plt.contour(xx, yy, zz, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
    plt.show()

plotDecisionBoundary()
```
