Import the libraries

```python
import scipy.io as s
import numpy as np
import matplotlib.pyplot as plt
```

Read the data

```python
data = s.loadmat('data.mat') # ['__header__', '__version__', '__globals__', 'X', 'Xval', 'yval']

X = data['X']
Xval = data['Xval']
yval = data['yval']
```

Plot the data

```python
plt.scatter(X[:, 0], X[:, 1])
plt.grid()
plt.show()
```

Find outliers and color them in red. You need to choose your own epsilon

```python
def p(x):
    mu = np.mean(x, axis=0) # np.sum(x, axis=0)/len(x)
    diag = np.diag(np.sum((x - mu)**2, axis=0)/len(x))
    return np.exp(-0.5* np.sum((x - mu) @ np.linalg.pinv(diag) * (x - mu), axis=1)) / ((2*np.pi)**(len(mu)/2)*(np.linalg.det(diag)**0.5))

pXval = p(Xval)

from sklearn.metrics import f1_score
eps = pXval[pXval <= pXval.mean()]
scores = []
for i in range(len(eps)):
    scores.append(f1_score(pXval <= eps[i], yval))

pX = p(X)
normal  = X[ pX >= eps[np.array(scores).argmax()] ]
anomaly = X[ pX <  eps[np.array(scores).argmax()] ]

plt.scatter( normal[:, 0],  normal[:, 1])
plt.scatter(anomaly[:, 0], anomaly[:, 1], c='r' )
plt.grid()
plt.show()
```
